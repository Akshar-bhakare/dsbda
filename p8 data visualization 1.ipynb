Data Visualization I
1. Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information
about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to
see if we can find any patterns in the data.
2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger
is distributed by plotting a histogram.

```python
import pandas as pd
import seaborn as sns # Seaborn is a data visualization library built on matplotlib
import matplotlib.pyplot as plt
```

```python
# Load the inbuilt 'titanic' dataset directly from seaborn
df = sns.load_dataset('titanic')
df.head()
```

```python
# Checking for basic info to see available columns and data types
df.info()
```

```python
# Finding patterns: Survival rate by sex using a countplot
# This helps us see if males or females had a higher survival count
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Survival count by Gender')
plt.show()
```

```python
# Finding patterns: Survival rate by passenger class
# This helps us see if passenger class (1st, 2nd, 3rd) affected survival
sns.countplot(x='survived', hue='pclass', data=df)
plt.title('Survival count by Passenger Class')
plt.show()
```

```python
# Plotting the distribution of ticket prices ('fare') using a histogram
# histplot automatically creates bins and counts the frequency of values
# kde=True adds a kernel density estimate curve to show the smooth distribution shape
sns.histplot(df['fare'], bins=30, kde=True, color='blue') 
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
