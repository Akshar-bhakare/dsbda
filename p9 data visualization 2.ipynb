1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution
of age with respect to each gender along with the information about whether they survived
or not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the above statistics.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the inbuilt 'titanic' dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# Plotting a box plot for distribution of 'age' with respect to 'sex' 
# We use 'hue' parameter to split the boxes by 'survived' status (0 = No, 1 = Yes)
sns.boxplot(x='sex', y='age', hue='survived', data=df)
plt.title('Distribution of Age by Gender and Survival Status')
plt.show()
```

### Observations from the Box Plot:
1. **Age Distribution**: The median age of male passengers (both survived and not survived) is slightly higher than that of female passengers.
2. **Survival in Young Males**: Among males, a larger portion of younger individuals (children/teens) survived compared to older ones. This is visible by the lower median and interquartile range of the 'survived' boxplot for males.
3. **Outliers**: There are several outliers (represented by dots above the whiskers) in both the male and female categories, which represent older passengers aboard the ship.
4. **Overall Survival Spread**: For females, the age distribution between those who survived and those who did not is quite similar, though younger females slightly trend toward survival.
