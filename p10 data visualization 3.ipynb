Download the Iris flower dataset or any other dataset into a DataFrame. (e.g.,
https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and give the inference as:
1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a boxplot for each feature in the dataset.
4. Compare distributions and identify outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
# We can use seaborn's inbuilt iris dataset to avoid downloading manually
df = sns.load_dataset('iris')
df.head()
```

```python
# 1. List down features and their types
print("Dataset Features and their Data Types:")
print(df.dtypes)

# Observation on types:
# - sepal_length, sepal_width, petal_length, petal_width are Numeric (float64).
# - species is Nominal (object/string).
```

```python
# 2. Create a histogram for each numeric feature
# We use plt.subplots to display multiple graphs in a single figure grid (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 8)) 

sns.histplot(df['sepal_length'], ax=axes[0, 0], kde=True)
sns.histplot(df['sepal_width'], ax=axes[0, 1], kde=True)
sns.histplot(df['petal_length'], ax=axes[1, 0], kde=True)
sns.histplot(df['petal_width'], ax=axes[1, 1], kde=True)

plt.tight_layout() # Adjust layout so titles and labels don't overlap
plt.show()
```

```python
# 3. Create a boxplot for each numeric feature to identify outliers
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Boxplots visually display the median, quartiles, and points that fall outside the typical range (outliers)
sns.boxplot(y=df['sepal_length'], ax=axes[0, 0])
sns.boxplot(y=df['sepal_width'], ax=axes[0, 1])
sns.boxplot(y=df['petal_length'], ax=axes[1, 0])
sns.boxplot(y=df['petal_width'], ax=axes[1, 1])

plt.tight_layout()
plt.show()
```

### 4. Compare distributions and identify outliers
- **Sepal Length**: The histogram shows a fairly normal distribution. The boxplot shows there are **no outliers**.
- **Sepal Width**: The histogram distribution is roughly bell-shaped (normal). The boxplot clearly shows a few **outliers** (dots outside the whiskers) on both the upper and lower extremes.
- **Petal Length**: The histogram shows a bimodal distribution (two peaks, likely corresponding to different species). **No outliers** are present in the boxplot.
- **Petal Width**: The histogram is also slightly bimodal. **No outliers** are detected in the boxplot.
