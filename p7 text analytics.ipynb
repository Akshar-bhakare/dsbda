Text Analytics
1. Extract Sample document and apply following document preprocessing methods:
Tokenization, POS Tagging, stop words removal, Stemming and Lemmatization.
2. Create representation of document by calculating Term Frequency and Inverse Document
Frequency.

```python
import nltk # Natural Language Toolkit, standard library for NLP in Python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer # Used to calculate TF-IDF

# Download necessary NLTK data files (only need to run once per environment)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
# Sample document text
text = "Natural language processing (NLP) is a machine learning technology that gives computers the ability to interpret, manipulate, and comprehend human language."
```

```python
# 1. Tokenization
# Splitting the text into individual words or punctuation marks
words = word_tokenize(text)
print("Tokenized Words:\n", words)
```

```python
# 2. POS Tagging (Part of Speech)
# Identifying the grammatical group of each word (noun, verb, adjective, etc.)
pos_tags = pos_tag(words)
print("\nPOS Tags:\n", pos_tags)
```

```python
# 3. Stop words removal
# Removing common words that add no significant meaning (like 'is', 'a', 'the')
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if not w.lower() in stop_words]
print("\nAfter Stop Words Removal:\n", filtered_words)
```

```python
# 4. Stemming
# Reducing words to their root/base form by chopping off suffixes (e.g., 'manipulate' -> 'manipul')
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(w) for w in filtered_words]
print("\nStemmed Words:\n", stemmed_words)
```

```python
# 5. Lemmatization
# Similar to stemming but reduces words to an actual dictionary word (lemma)
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]
print("\nLemmatized Words:\n", lemmatized_words)
```

```python
# 6. Term Frequency and Inverse Document Frequency (TF-IDF)
# Creating a list of sentences/documents
documents = [
    "Natural language processing (NLP) is a machine learning technology.",
    "Machine learning is a subset of artificial intelligence."
]
# Initialize the TF-IDF Vectorizer
tfidf = TfidfVectorizer()
# Fit and transform the documents into a TF-IDF matrix (a numeric representation)
result = tfidf.fit_transform(documents)

print("\nFeature Names (Vocabulary):\n", tfidf.get_feature_names_out())
print("\nTF-IDF Matrix (Rows are documents, Columns are words):\n", result.toarray())
```
