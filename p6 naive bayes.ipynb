```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Gaussian Naive Bayes is suitable for continuous data like Iris features
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
# Load the dataset
df = pd.read_csv('/content/Iris.csv') # Assuming the dataset is uploaded to colab as in previous practicals
df.head()
```

```python
# Separate features (X) and target variable (y)
# We take all columns except 'Id' and 'Species' as features
x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']] 
y = df['Species'] # Target variable to predict
```

```python
# Split data into training and testing sets (75% training, 25% testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
# Initialize the Naive Bayes Classifier
model = GaussianNB()
# Train the model using the training data
model.fit(x_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = model.predict(x_test)
```

```python
# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
# For multiclass, TN/FP/FN/TP is usually computed per class or micro/macro averaged.
# We'll print overall accuracy, precision, and recall.
accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
# Using macro average since it's a multi-class classification problem (Iris has 3 classes)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')

print(f"Accuracy: {accuracy}")
print(f"Error Rate: {error_rate}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
```
